# For Loop

```python
# 1.Reverse a string without slicing
string="Pentagon Space"
rev=''
for i in string:
    rev=i+rev
rev
```

```python
# 2.Reverse starting 5 index from string without slicing
string="Mastering the future"
rev=''
num=0
for i in string:
    if num<6:
        rev=i+rev
        num+=1
    else:
        rev=rev+i
        num+=1
print(rev)
```

```python
# 3. Program to remove duplicate values from string
string="Pentagon Space"
unique=''
for i in string:
    if i not in unique:
        unique=unique+i
unique
```

# While Loop

```python
# 4.Reverse a string without slicing
name="Pentagon Space"
i=0
rev=''
while(i<len(name)):
    rev=name[i]+rev
    i+=1
print(rev)
```

```python
# 5.Reverse last 5 index from string without slicing
string="Pentagon Space"
i=0
last_index=len(string)-5
rev=''
while(i<len(string)):
    if i>last_index:
        rev=string[i]+rev
        i=i+1
    else:
        rev=rev+string[i]
        i=i+1
print(rev)
```

```python
# 6. Program to remove duplicate values from string
string="Pentagon Space"
unique=''
i=0
while(i<len(string)):
    if string[i] not in unique:
        unique=unique+string[i]
    i+=1
print(unique)
```
