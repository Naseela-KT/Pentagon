```python
# 1.Check whether area of a rectangle greater than perimeter of a rectangle
l=5
b=10
area=l*b
perimeter=2*(l+b)
if(area>perimeter):
    print("Area is greater than Perimeter")
else:
    print("Area is less than Perimeter")
```

```python
# 2.From the age of 3 people determine the oldest person among them
p1=25
p2=34
p3=45
if(p1>p2):
    if(p1>p3):
        print("First person is oldest")
    else:
        print("Third person is oldest")
elif(p2>p3):
    print("Second person is oldest")
else:
    print("Third person is oldest")
```

```python
# 3. Program to divide them into 3 groups, based on users age
age=int(input("Enter your age: "))
if(age<18):
    print("Minors who are not eligible to work")
elif(age>=18 and age<=60):
    print("Eligible to work")
else:
    print("Too old to work as per govt. regulations")
```

```python
# 4. To check whether an alphabet is a vowel or not
ch=input("Enter a character: ")
if(ch=="a" or ch=="A" or ch=="e" or ch=="E" or ch=="i" or ch=="I" or ch=="o" or ch=="O" or ch=="u" or ch=="U"):
    print("Vowel")
else:
    print("Consonant")
```
